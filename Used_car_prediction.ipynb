```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head(1)
```

```python
df_test.head(1)
```

```python
df_train.drop('id',axis=1,inplace=True)
```

```python
df_test.drop('id',axis=1,inplace=True)
```

```python
df_test.isna().sum() * 100 /len(df_train)
```

```python
def fill_na(column_name, d_type):
    if d_type == 'train':
        elements = df_train[i].dropna().unique()
        proba = [df_train[column_name].value_counts()[i] / len(df_train[column_name].dropna()) for i in elements]
        df_train.fillna(np.random.choice(a=elements,p=proba),inplace=True)
    else:
        elements = df_test[i].dropna().unique()
        proba = [df_test[column_name].value_counts()[i] / len(df_test[column_name].dropna()) for i in elements]
        df_test.fillna(np.random.choice(a=elements,p=proba),inplace=True)
```

```python
for i in df_train.isna().sum()[df_train.isna().sum() > 0].index:
    fill_na(i,'train')
```

```python
for i in df_test.isna().sum()[df_test.isna().sum() > 0].index:
    fill_na(i,'test')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df_test.columns:
    if df_test[i].dtype ==object:
        if len(set(set(df_test[i])).difference(set(df_train[i]))) > 0:
            print(i)
            print(set(set(df_test[i])).difference(set(df_train[i])))
```

```python
df_train['model'].value_counts()
```

```python
df_test.replace('allroad 2.7T','F-150 XLT',inplace=True)
```

```python
df_train['engine'].value_counts()
```

```python
df_test['engine'].replace('177.0HP 1.6L 4 Cylinder Engine Plug-In Electric/Gas','355.0HP 5.3L 8 Cylinder Engine Gasoline Fuel',inplace=True)
```

```python
for i in df_test.columns:
    if df_test[i].dtype == object:
        label = LabelEncoder()
        df_train[i] = label.fit_transform(df_train[i])
        df_test[i] = label.transform(df_test[i])
```

```python
df_train.corr()
```

```python
df_train.drop('brand',axis=1,inplace=True)
df_test.drop('brand',axis=1,inplace=True)
```

```python
df_train.drop('model',axis=1,inplace=True)
df_test.drop('model',axis=1,inplace=True)
```

```python
df_train.drop('fuel_type',axis=1,inplace=True)
df_test.drop('fuel_type',axis=1,inplace=True)
```

```python
df_train.drop('transmission',axis=1,inplace=True)
df_test.drop('transmission',axis=1,inplace=True)
```

```python
df_train.drop('ext_col',axis=1,inplace=True)
df_test.drop('ext_col',axis=1,inplace=True)
```

```python
df_train.drop('int_col',axis=1,inplace=True)
df_test.drop('int_col',axis=1,inplace=True)
```

```python
df_train.drop('clean_title',axis=1,inplace=True)
df_test.drop('clean_title',axis=1,inplace=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'criterion' : ["squared_error", "friedman_mse","poisson"],
    'splitter' : ['best','random'],
    'max_depth' : [2,4,6,8,10,None]
}
```

```python
final_model = GridSearchCV(model,param_grid=params,cv=3,verbose=2)
```

```python
X_train = df_train.drop('price',axis=1)
X_test = df_test.copy()
y_train = df_train['price']
```

```python
final_model.fit(X_train,y_train)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['price'] = y_pred
```

```python
df_sub.to_csv('sub.csv',index=False,index_label=False)
```
